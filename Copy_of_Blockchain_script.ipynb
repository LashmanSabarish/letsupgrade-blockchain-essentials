```python
from IPython.display import clear_output
from dict_hash import sha256
import time
```

```python
def mineBlock(ditBlock):
    flag = True
    nonce = 0
    start = time.time()
    while flag:
        dit["nonce"] = nonce
        hsh = sha256(ditBlock)
        pos1 = hsh[0]
        pos2 = hsh[1]
        pos3 = hsh[2]
        pos4 = hsh[3]
        pos5 = hsh[4]

        if pos1 == '0' and pos2 == "0" and pos3 == "0" and pos4 == "0" and pos5 == "0"  :
            flag = False
            dit["hash"] = hsh
            print(ditBlock)
            end = time.time()
            print(f"Runtime of the program is {end - start}")
        else:
            print(nonce)
#             clear_output(wait=False)
            nonce = nonce + 1
```
